```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_classification
#creating datasets
```

```python
## create the dataset
X ,y = make_classification(n_samples=1000, n_features=10, n_classes=2, random_state=15)
#10 independent feature
#2 outputs(n_classes)
pd.DataFrame(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split( X,y,test_size=0.30,random_state=42)
```

```python
#model training
from sklearn.linear_model import LogisticRegression
logistic=LogisticRegression()
logistic.fit(X_train,y_train)
y_pred=logistic.predict(X_test)
print(y_pred)
```

```python
logistic.predict_proba(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
score=accuracy_score(y_test,y_pred)
print(score)
print(classification_report(y_pred,y_test))
print(confusion_matrix(y_pred,y_test))
```

### HyperParameter tuning and CV


# GRIDSEARCH CV

```python
model=LogisticRegression()
penalty=['l1', 'l2', 'elasticnet']  #ridge lasso elastic net
c_values=[100,10,1.0,0.1,0.01]
solver=['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'] #algoeithim for optimization of problems 'liblinear for small datasets. 'sag and saga for large ones.  
```

```python
#convertig to key vallue pair
dict(penalty=penalty,C=c_values,solver=solver)
```

```python
params=dict(penalty=penalty,C=c_values,solver=solver)
```

```python
from sklearn.model_selection import StratifiedKFold
cv=StratifiedKFold()
```

```python
## GridSearchCV
from sklearn.model_selection import GridSearchCV

grid=GridSearchCV(estimator=model,param_grid=params,scoring='accuracy',cv=cv,n_jobs=-1)
```

```python
grid
```

```python
# this will find the best parameters for our model
grid.fit(X_train,y_train)
```

```python
#to see the best parameters
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred=grid.predict(X_test)
```

```python
score=accuracy_score(y_pred,y_test)
print(score)
print(classification_report(y_pred,y_test))
print(confusion_matrix(y_pred,y_test))
```

# Randomized SeaechCv 
takes less time than gridparam as it takes sample of our model and train the model.
it picks random parameters and random attributes

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
model=LogisticRegression()
randomcv=RandomizedSearchCV(estimator=model,param_distributions=params,cv=5,scoring='accuracy')
```

```python
randomcv.fit(X_train,y_train)
```

```python
randomcv.best_score_
```

```python
randomcv.best_params_
```

```python
y_pred=randomcv.predict(X_test)
```

```python
score=accuracy_score(y_pred,y_test)
print(score)
print(classification_report(y_pred,y_test))
print(confusion_matrix(y_pred,y_test))
```
